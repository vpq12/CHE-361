## fitting an empirical model

we derived the following ODE model in class for the thermistor temperature $T(t)$ [the output] in response to the external temperature $T_e(t)$ imposed [the input].

\begin{equation}
\frac{C}{UA}\frac{dT}{dt} + T = T_e.
\end{equation}

* $U$: heat transfer coefficient
* $A$: surface area
* $C$: thermal capacity

thus this is a first order process with time constant $\tau:=C/UA$.

i.e. the transfer function for this process is:

$$G(s)=\frac{\mathcal{T}^*(s)}{\mathcal{T}^*_e(s)}=\frac{1}{\tau s +1}$$

(see lecture notes on Canvas "thermistor dynamics")

**the goal** of this studio is to use the data to emperically identify the time constant $\tau$ of this process. this is in contrast to measuring $C$, $U$, and $A$ of the thermistor separately.

(4) given a guess for the time constant $\tau$, write a function `T_model(t, τ)` that returns the temperature according to the first order model for $T(t)$ and specified time constant $\tau$. you will need an `if` statement for when `t<0` and `t>0`. (see [docs](https://docs.julialang.org/en/v1/manual/control-flow/#man-conditional-evaluation-1))

```julia
T₀=T_Grant
T₁= T_ambient
τ=10
#τ=c/UA
function T_model(t,τ)
if t>0
    return T₀+(T₁-T₀)*(1-exp(-t/τ))
    
else 
    return T_Grant
end
end
plot(t,T_model.(t,τ))
```

```julia
T_model.(2,2)
```

(5) write a function `cost(τ)` that computes the cost function for a given guess for `τ`. as the "cost", use the sum of square errors between the predicted temperature and actual temperature. we seek to minimize the cost to determine the time constant `τ`. 

```julia
function cost(τ)
loss = 0
for i = 1:length(t)
residual = (T[i]- T_model(t[i],τ))
loss = loss + residual^2
end
return loss
end
```

(6) plot the cost function (y-axis: cost, x-axis: $\tau$) for a relevant interval of $\tau$'s so that you can clearly see the minimum. I'm not telling you the interval so you can think carefully about what scale you need for $\tau$ based on your plotted data. hint: what would you guess $\tau$ to be by eye-balling the data?

```julia
τ = collect(1.0:0.01:20.0)

plot(τ,cost.(τ))
```

(7) use the `optimize` function in `Optim.jl` to minimize the cost function you wrote. assign the optimum value of $\tau$ as a variable `τ_opt`. what is the optimal value of $\tau$ that best fits the data? what are the units?

see [Optim.jl docs](https://julianlsolvers.github.io/Optim.jl/stable/#user/minimization/#minimizing-a-univariate-function-on-a-bounded-interval) on minimizing a uni-variate function if you are struggling.

```julia
τ_opt = optimize(cost, 0.0, 0.50).minimizer # tell Optim.jl the optimum a is between 0 and 0.50
```

```julia
figure()
plot(t,T , color = "blue")
xlabel("time")
ylabel("temperature")
axhline(y = T_ambient, color= "yellow", label = "t_ambient")
axhline(y = T_Grant, color = "black" ,label = "t_grant")
plot(t, T_model.(t,τ_opt), color = "pink", label ="topt")
legend()
```

the reason is that, now that we have an empirical model for heat transfer to the thermistor, we can use the empirical model to simulate the response of the thermistor to *different* inputs. here, *different* means different from the input this data corresponds to.

an alternative way of thinking is that we identified $\tau=C/(UA)$ from the process response curve. we identified these three parameters lumped together, but none individually. since it's *this* quantity that determines the dynamics, I'm happy with that!

## putting the empirical model to use

(9) plot the response $T^*(t)$ to a ramp input in temperature, where the air is warmed by a heater at rate $0.01$ K/s. simulate the response for two minutes.

```julia
using Controlz
a = 0.01 
function T_star(t,τ)
if t<0
return 0
else
return a*t + a*τ*((exp(-t/τ))-1)
end
end
t = collect(0.0:0.01:120.0)
plot(t,T_star.(t,τ_opt))
ylabel("temperature")
xlabel("time")
```
