# how to numerically solve systems of non-linear equations

e.g. solve the following system of two non-linear equations for $x_1$ and $x_2$.

$$x_1-x_2+1=0$$
$$x_2-x_1^2-1=0$$

first, let's define two functions $f_1(x_1, x_2)$ and $f_2(x_1, x_2)$ that we want to be zero.

$$f_1(x_1, x_2) :=x_1-x_2+1$$

$$f_2(x_1, x_2) := x_2-x_1^2-1$$

```julia
f₁(x₁, x₂) = x₁ - x₂ + 1 # e.g. you can call f(2.0, 3.0) to return the value at x₁=2, x₂=3

f₂(x₁, x₂) = x₂ - x₁ ^ 2 - 1
```

next, load in the `NLSolve` package. `NL` stands for "NonLinear".

```julia
using NLsolve
```

now, define a function `make_zero!` that takes in a vector $\mathbf{x}=[x_1, x_2]$ and a vector $\mathbf{F}=\mathbf{F}(\mathbf{x})=[f_1(\mathbf{x}), f_2(\mathbf{x})]$ and populates the entries of $\mathbf{F}$. we aim to find the vector $\mathbf{x}_0$ such that $\mathbf{F}(\mathbf{x}_0)=\mathbf{0}$, since this implies that $f_1(x_1, x_2)=f_2(x_1,x_2)=0$.

(another way of writing $f_1(x_1, x_2)$ in vector form is $f_1(\mathbf{x})$.

a little awkward to program, but this is the form that `NLsolve` wants.

```julia
function make_zero!(F, x)
    # x = [x₁, x₂]. unpack the vector to make it clear what the entries represent
    x₁ = x[1]
    x₂ = x[2]
    
    # put the two functions you want to be zero as the two entries of F = [f₁, f₂]
    F[1] = f₁(x₁, x₂)
    F[2] = f₂(x₁, x₂)
end
```

finally, pass this function `make_zero!` to `nlsolve` with an intial guess for the solution. `nlsolve` will find the zero $\mathbf{x}_0$ such that $\mathbf{F}(\mathbf{x}_0)=\mathbf{0}$.

```julia
x₀_guess = [2.0, 4.0]
res = nlsolve(make_zero!, x₀_guess)
```

extract the zero of $\mathbf{F}(\mathbf{x})=[f_1(\mathbf{x}), f_2(\mathbf{x})]$ from the result `res`.

```julia
x₁, x₂ = res.zero
```

sanity check

```julia
f₁(x₁, x₂)# close to zero, to numerical precision
```

```julia
f₂(x₁, x₂) # close to zero, to numerical precision
```

note that if you use a different intial guess, the algorithm finds a different zero of $\mathbf{F}(\mathbf{x})=[f_1(\mathbf{x}), f_2(\mathbf{x})]$.

```julia
x₀_guess = [0.0, 0.0]
nlsolve(make_zero!, x₀_guess).zero
```

# how to numerically differentiate a function

e.g. consider
$f(x_1, x_2)= x_1 x_2 + x_2^2 + 4$

we seek the partial derivatives evaluated at certain values of $x_1$ and $x_2$: $x_1=\bar{x_1}=2$ and $x_2=\bar{x_2}=4$.

find:
* $\dfrac{\partial f}{\partial x_1}\bigg|_{x_1=2, x_2=4}$
* $\dfrac{\partial f}{\partial x_2}\bigg|_{x_1=2, x_2=4}$

let's use the `Calculus` package to numerically differentiate $f(x_1, x_2)$ instead of doing the math by hand (though easy in this case).

```julia
using Calculus: gradient
```

define $f(x_1, x_2)$ as a function of a vector $\mathbf{x}=[x_1, x_2]$. we need to define `f` this way for `Calculus` to be able to take the derivative.

```julia
function f(x) # x is a vector
    # x = [x₁, x₂]. unpack the vector for clarity
    x₁ = x[1]
    x₂ = x[2]
    return x₁ * x₂ + x₂ ^ 2 + 4
end

x̄ = [2.0, 4.0] # a vector. so x₁ = 2.0, x₂ = 4.0
f(x̄)
```

now you can compute the gradient $\dfrac{\partial f}{\partial \mathbf{x}}$ (evaluated at $\mathbf{x}=\bar{\mathbf{x}}$) by passing the function (of a vector), `f`, to `gradient`.

`gradient` returns a vector. the first entry is $\dfrac{\partial f}{\partial x_1}$ and the second entry is $\dfrac{\partial f}{\partial x_2}$, both evaluated at the `x̄` vector passed to `gradient` as the second argument.

```julia
gradient(f, x̄) # [∂f/∂x₁, ∂f/∂x₂] evaluated at x̄
```

sanity check (we can do these partial derivatives in our heads!).
there is some numerical error, but not much.

```julia
∂f_∂x₁ = x̄[2]
```

```julia
∂f_∂x₂ = x̄[1] + 2 * x̄[2]
```

# bioreactor: continuous cultivation of micro-organisms



* $F=F(t)$ [L/hr]: incoming volumetric flow rate of broth containing substrate (but not micro-organisms)
* $x=x(t)$ [g micro-organisms/L]: concentration of micro-organisms in the well-mixed bioreactor and thus also in the exit stream
* $s=s(t)$ [g substrate/L]: concentration of (unconsumed) substrate in the well-mixed bioreactor and thus also in the exit stream

A mixed flow bioreactor system has a continuous feed and product stream, and the volume of liquid in the tank is constant. The incoming feed is sterile ($x_f=0$). The dynamic model for the bioreactor is comprised of two first-order, coupled, non-linear ODEs derived from a mass balance on micro-organisms and a mass balance on substrate. For a complete description of parameters see the lecture notes on the bioreactor. 

\begin{align}
    \frac{dx}{dt} &= - \frac{Fx}{V}+ \mu_{max}\Big(\frac{s}{K_S+s}\Big)x  =:f_x(x,s,F)\\
    \frac{ds}{dt} &= \frac{F}{V}(s_f-s) - Mx-\mu_{max}\Big(\frac{s}{K_S+s}\Big)\frac{x}{Y} =: f_s(x,s,F)
\end{align}   
    
The bioreactor has **one input**, $F$, and **two (coupled) outputs**, $x$ and $s$.

The parameters for the model are below.

```julia
M = 0.2 # g substrate to keep micro-organisms alive / [g micro-organisms - hr]
μ_max = 2.88 # 1 / hr
Y = 0.2 # g micro-organisms / g substrate to produce micro-organisms
Kₛ = 4.0 # g substrate / L
V = 1.0 # L
sf = 7 # g substrate / L
```

(0) define two Julia functions, `fₓ(x, s, F)` and `fₛ(x, s, F)` that correspond to the right-hand-sides of the set of coupled ODEs governing $x(t)$ and $s(t)$, respectively.

```julia
f_x(x_1,s,F) = -((F*x_1)/V )+μ_max*(s/(Kₛ+s))*x_1

f_s(x_1,s,F) = ((F*(sf-s))/V)-(M*x_1)-(μ_max*x_1*s)/(Y*(Kₛ+s))
```

though $F=F(t)$, at steady state conditions, the value of the feed flow rate is $\bar{F}$ below. we decorate $F$ with a bar to indicate that it is a steady state value.
