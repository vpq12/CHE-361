# second order (SO) transfer functions

a second order transfer function $G(s)$ relating an input $u$ to an output $y$ takes the form:
\begin{equation}
    G(s)= \frac{Y^*(s)}{U^*(s)} = \frac{K}{\tau^2 s^2 + 2\tau \xi s + 1}
\end{equation}
